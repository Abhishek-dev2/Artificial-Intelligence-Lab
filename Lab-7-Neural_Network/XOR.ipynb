```python
import torch
from torch import optim
import sklearn
```

```python
# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
# N, D_in, H, D_out = 12, 4, 10, 1
D_in, H, D_out = 10, 100, 1
```

```python
def n_input(n):
    total = 2**n
    X = []
    Y = []
    for i in range(total):
        temp = []
        x = i
        pos = 0
        ans = 0 
        while x > 0:
            if x&1 : 
                temp.append(1)
                ans ^= 1
            else : 
                temp.append(0)
            x //= 2
        while len(temp) < n:
            temp.append(0)
        X.append(temp)
        Y.append([ans])
    return X, Y
```

```python
X,y = n_input(10)
```

```python
print(X[1])
print(y[1])
```

```python
split_index = int(len(X)*0.8)
X_train = torch.cuda.FloatTensor(X[:split_index])
X_test = torch.cuda.FloatTensor(X[split_index:])
y_train = torch.cuda.FloatTensor(y[:split_index])
# y_train.view(-1, 1)
y_test = torch.cuda.FloatTensor(y[split_index:])
```

```python
print(len(X_train))
print(y_train.size())
# print(X_train)
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(D_in, H),
    torch.nn.ReLU(),
    torch.nn.Linear(H, D_out),
)
```

```python
model.cuda(0)
```

```python
loss_fn = torch.nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.01, momentum=0.9)
```

```python
learning_rate = 1e-4
best_loss = 1000
for t in range(2500):

    optimizer.zero_grad()
    y_pred = model(X_train)

    loss = loss_fn(y_pred, y_train)
    if loss < best_loss:
        best_loss = loss
        torch.save(model, './model')
        print(t, loss.item())    
#     print(t, loss.item())

    # Zero the gradients before running the backward pass.
#     model.zero_grad()
    
    loss.backward()
    optimizer.step()

#     with torch.no_grad():
#         for param in model.parameters():
#             param -= learning_rate * param.grad
```

```python
model = torch.load('./model')
```

```python
pred = model(X_test).cpu().detach().numpy()
```
